```python
import requests
import json

# Ethereum node URL
node_url = "http://localhost:8545"

# List of token tickers on your watchlist
tickers = ["ETH", "DAI", "USDT"]

# Example total supply for each token (you would need to fetch these dynamically)
total_supplies = {
    "ETH": 10000000000000,
    "DAI": 10000000000000,
    "USDT": 1000000000000
}

# Transaction hashes to check
transaction_hashes = [
    "0x1234567890abcdef1234567890abcdef1234567890abcdef1234567890abcdef",
    "0xabcdef1234567890abcdef1234567890abcdef1234567890abcdef1234567890"
]

# Percentage threshold
percentage_threshold = 0.01 # 1%

for ticker, total_supply in total_supplies.items():
    for tx_hash in transaction_hashes:
        # Fetch transaction details
        response = requests.post(
            node_url,
            json={
                "jsonrpc": "2.0",
                "method": "eth_getTransactionByHash",
                "params": [tx_hash],
                "id": 1
            }
        )
        tx_data = response.json()
        if 'result' in tx_data:
            tx_value = int(tx_data['result']['value'], 16)
            percentage = (tx_value / total_supply) * 100
            if percentage > percentage_threshold:
                print(f"Transaction {tx_hash} for {ticker} exceeds {percentage_threshold}% of total supply.")
```
